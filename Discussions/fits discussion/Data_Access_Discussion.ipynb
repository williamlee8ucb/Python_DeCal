```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
from astropy.table import QTable
```

```python
fits_file = fits.open('fits_file_disc(2).fits')
# creates an HDUList object

fits_file.info()
```

```python
fits_file[1]??
```

```python
data = fits_file[1].data
# in this case, is a way to access parameters of object ".parameter"
fits_qt = QTable(data)
fits_qt

# can also do this with pandas dataframe
```

```python
wavelength = fits_qt["Wavelength (Angstroms)"]
flux = fits_qt["Flux"]
```

```python
plt.figure(figsize = (25, 6))

plt.plot(wavelength, flux)

plt.xlabel("Wavelength (Angstroms)")
plt.ylabel("Flux (erg/s/cm2)")
plt.title("Sample Spectra")
plt.grid()

plt.show()
```

### Question 1: Pillars of Creation!

In this question, we're going to visualize real world data using ImShow. Specifically, we'll be plotting data that represents the Pillars of Creation (Eagle Nebula/Messier 16). Things that you will have to do...
- Load in the data from the provided CSV file (you will have to use numpy or pandas)
- Plot the data using ImShow
- Pick an appropriate colormap, and show the colorbar
- Use the vmin/vmax parameters to enhance the colors of the image
- Set an appropriate origin and extent (you can also use the xlim parameter)

```python
# Create your plot here
image_data = np.loadtxt("astro_image.csv")
image_data
# pandas not used because np.loadtxt loads in data as numpy arrays
# easier to find min and max

plt.figure()
plt.imshow(image_data, vmin = 50, vmax = 300, extent = [0, 1550, 0, 1400])
# need to assess array to see minimum and maximum and outliers 
# extent creates borders
plt.set_cmap("bone")
plt.colorbar()
plt.title("Pillars of Creation")
plt.show()
```

### Question 2: Statistical Plots

In this question, we're going to develop upon some of the skills we learned about reading documentation and plotting using Seaborn. You can use the seaborn and matplotlib documentation pages to help you with this assignment!
Using the 'baby' dataset- let's make some plots that illustrate key features of the data.

- First- load in the dataset (should be called baby.csv). You should use pandas to do this.
- A Count Plot that contains information of whether the mother was a maternal smoker or not
- A Scatter plot with the Maternal Height and Birth Weight, and separate colors to depict whether the mother was a smoker or not.
- A histogram of the distribution of baby birth weights. Because of the range of weights, you can use 20 bins.
- 2 Side by side box plots of birth weights (one for babies whose mothers were maternal smokers, and one for those whose mothers weren't maternal smokers)

```python
#Load in the Dataset here!
babies = pd.read_csv("baby.csv")
babies
```

```python
#Count Plot
plt.figure()
sns.countplot(babies, x = "Maternal Smoker")
```

```python
#Scatter Plot
sns.scatterplot(babies, x = "Maternal Height", y = "Birth Weight", hue = "Maternal Smoker")
```

```python
#Histogram
sns.histplot(babies, x = "Birth Weight", bins = 20)
```

```python
#Box Plots
sns.boxplot(data = babies, x = "Birth Weight", hue = "Maternal Smoker")
```

### Question 3: Scaling Plots

In this question, we're going to see how changing the scale on a plot will affect the way we view and analyze data.
To do this, we're going to use some generated data.
- Create a figure with 2 subplots.
- On the right side, plot the data (scaling_data). Do you notice something about the y-axis scale?
- On the left side, plot this same data with a readjusted y-axis scale. (you can use a matplotlib operation, or manually rescale the data, and then plot it)

```python
#Data
scaling_data = [10**i for i in range(5)]
```

```python
#Create your plots here!
plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 4))
ax2.plot(scaling_data)
ax1.plot(scaling_data)
ax1.set_yscale('log')

ax2.set_title("Scaling Data")
ax1.set_title("Log Scaled Scaling Data")
```

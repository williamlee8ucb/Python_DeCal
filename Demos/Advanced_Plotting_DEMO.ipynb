```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## SUBPLOTS

#Generating some random data stuff, not important
x = np.linspace(0,2*np.pi, 100)
x_1 = np.linspace(0,100,100)
y1 = 2*np.sin(x)*np.exp(-x)
y2 = np.exp(x/4)
x_scatter = np.linspace(0,100,100)
y1_scatter = (x_scatter) + np.random.randint(low=-50, high=50, size=100)
y2_scatter = x_scatter + np.cos(x)*np.random.random(100)*100

#one figure with 2 columns, and 2 rows
f, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, figsize=(10,10))

#top left

ax1.plot(x, y1)
ax1.set_title("X and Y1")
ax1.set_xlabel("X-Axis")
ax1.set_ylabel("Y-Axis")

#top right

ax2.scatter(x_scatter, y1_scatter)
ax2.set_title("X Scatter and Y1 Scatter")
ax2.set_xlabel("X-Axis")
ax2.set_ylabel("Y-Axis")

#bottom left
ax3.scatter(x, y2)
ax3.set_title("X and Y2")
ax3.set_xlabel("X-Axis")
ax3.set_ylabel("Y-Axis")

#bottom right
ax4.scatter(x_scatter, y2_scatter)
ax4.set_title("X Scatter and Y2 Scatter")
ax4.set_xlabel("X-Axis")
ax4.set_ylabel("Y-Axis")

f.tight_layout()
plt.show()
```

```python
#3D Functions

x,y=np.meshgrid(np.linspace(-1,1,100),np.linspace(-2,2,100)) # makes a grid from given set of x and y values
f = x + y

plt.figure()
plt.imshow(f, origin = "lower") # can change color map with "cmap = " argument
plt.colorbar()
plt.show()
# shows sum of x and y (represented as a color gradient) at each value of x and y
```

```python
# DIY
x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5,5,100))
f = np.sqrt(x**2 + y**2)

plt.figure()
plt.imshow(f, origin = "lower")
plt.colorbar()
plt.show()
```

```python
## BAR PLOT
penguins = sns.load_dataset("penguins")
# seaborn, package allows for more statistical plots
penguins
```

```python
plt.figure()
sns.barplot(penguins, x = "island", y = "body_mass_g") # seaborn
plt.title("Penguin Homes vs Body Mass")
plt.xlabel("Penguin Home Islands")
plt.ylabel('Average Body Mass')
plt.show()
```

```python
# using hue parameter
# splits data based on a certain variable
plt.figure()
sns.barplot(penguins, x = "island", y = "body_mass_g", hue = "sex") # seaborn
# splits data into male and female penguins
plt.title("Penguin Homes vs Body Mass by Sex")
plt.xlabel("Penguin Home Islands")
plt.ylabel('Average Body Mass')
plt.show()
```

```python
## COUNT PLOT
titanic = pd.read_csv("titanic_data.csv")
titanic
```

```python
plt.figure()
ax = sns.countplot(titanic, x = "Pclass")
ax.bar_label(ax.containers[0], label_type = "edge")
# bar plot count label, from stackoverflow
plt.title("Passenger Class")
plt.xlabel("Class")
plt.ylabel('Number of Passengers')
plt.show()
```

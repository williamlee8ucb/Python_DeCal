```python
#SLICING

# creating a sample dataframe 
data2 = pd.DataFrame({'Brand': ['Maruti', 'Hyundai', 'Tata', 
                               'Mahindra', 'Maruti', 'Hyundai', 
                               'Renault', 'Tata', 'Maruti'], 
                     'Year': [2012, 2014, 2011, 2015, 2012, 
                              2016, 2014, 2018, 2019], 
                     'Kms Driven': [50000, 30000, 60000, 
                                    25000, 10000, 46000, 
                                    31000, 15000, 12000], 
                     'City': ['Gurgaon', 'Delhi', 'Mumbai', 
                              'Delhi', 'Mumbai', 'Delhi', 
                              'Mumbai', 'Chennai',  'Ghaziabad'], 
                     'Mileage':  [28, 27, 25, 26, 28, 
                                  29, 24, 21, 24]}) 
  
# displaying the DataFrame 
display(data2) 
```

```python
# Helpful link for following exercises: https://builtin.com/software-engineering-perspectives/pandas-iloc
```

```python
# LOC: LABEL BASED
# bold formatted stuff are labels
# format pd.loc[row, column]

# SINGLE POINT
one_point = data2.loc[1, "Year"]
print(one_point)

# RANGES OF ROWS/COLUMNS
range = data2.loc[0, :]
range1 = data2.loc[3:5, "City"]
# inclusive because they are labels, not indexes
print(range)
print(range1)

#CONDITIONALS
condition = data2.loc[(data2.Brand == "Maruti") & (data2.Mileage > 26)]
print(condition)
```

```python
# ILOC: INDEX BASED

# SPECIFIC ROWS, RANGES OF ROWS/COLUMNS
ind_range = data2.iloc[3:5, 0:3]
print(ind_range)
# not inclusive, since using indexes instead of labels
# is a series, not a dataframe; must be created as a new dataframe
```

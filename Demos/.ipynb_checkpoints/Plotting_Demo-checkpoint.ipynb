```python
x = np.linspace(0, 2*np.pi, 100) # creates 100 values (see documentation)
print(x)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure() # "canvas" for graph

plt.plot(x, y)
plt.plot(x, y2) # overlaid plots

plt.title("Example: Sine Function")
plt.xlabel("x")
plt.ylabel("y") # labelling

plt.legend(["Sine", "Cosine"], loc = "upper right") # names should be in order
                                                    # can change legend location (see docs)

plt.show()
```

```python
# CUSTOMIZING PLOTS DEMO

# documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

plt.figure() 

plt.plot(x, y, color = "magenta", label = "sine", linewidth = 1, alpha = .5, linestyle = "dashdot") 
# alpha is opacity, values from 0 to 1
plt.plot(x, y2, color = "lime", label = "cosine", marker = "3", linewidth = 45, alpha = 1) 
plt.plot(x, y2*2, color = "darkorange", label = "shittin' 'n' gigglin'", marker = (6, 0, 180), linewidth=30, alpha = .25)


plt.title("Example: Sine Function")
plt.xlabel("x") # can add as plot argument
plt.ylabel("y")

plt.legend(loc = "upper right") 

plt.show()
```

```python
# SUBPLOTS
# collection of smaller plots

fig, axs = plt.subplots(2,2) # rows, columns

axs[0,0].plot(x, y, color = "lavender", linewidth=20)
axs[0, 1].plot(x, y2)

plt.show()
```

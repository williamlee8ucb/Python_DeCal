```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1) How Many Plots?

# 1.1:
def sinusoid(x, A, w):
    return A*np.sin(w*x)

# 1.2:
x_data = np.linspace(0, 2*np.pi, 1000)

# 1.3:
A_data = np.linspace(1, 5, num = 5)
w_data = np.linspace(1, 2, num = 5)
A_data
```

```python
# 1.4 and 1.5:
plt.figure()

y_data = []
for i in range(5):
    y = sinusoid(x_data, A_data[i], w_data[i])
    y_data.append(y)
    plt.plot(x_data, y_data[i])

```

```python
# 2) Randomness

list1 = np.random.randint(0, 101, 40)
list2 = np.random.randint(0, 101, 40)
```

```python
plt.figure(figsize = [13, 8])

plt.plot(list1, color = 'orange', linewidth = 10)
plt.plot(list2, color = 'red', linestyle = 'dashed')

plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.legend(["Line 1", "Line 2"], loc = 'upper right')
```
